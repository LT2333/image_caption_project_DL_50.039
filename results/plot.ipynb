```python
def plot2(name, list):
    for i in range(len(list)):
        plt.plot(list[i])

    plt.title( name )
    plt.xlabel("epoch")
    plt.ylabel(name)
    plt.legend(["Model1", "Model2","Model3"], loc = "upper right")
    plt.savefig(name)
    plt.show()
    plt.clf()
```

```python
plot2("Avg Training Acuracy",[acc_1,acc_2,acc_3])
plot2("Avg Training Loss",[loss_1,loss_2,loss_3])
```
